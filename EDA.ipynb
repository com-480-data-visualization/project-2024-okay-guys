## A. Data Loading, Preparation and Cleaning

#### **Data Cleaning**

1. Replace the NaN values in the Medal column with "None" as they are not considered missing values but rather as separate pieces of information.

```python
olympics_data['Medal'] = olympics_data['Medal'].fillna('None')
display(olympics_data.head())
```

2. Drop uncessary column `Games` which gives information provided in the column `Year` and `Season`

```python
olympics_data.drop(columns=['Games'], inplace=True)
```

3. Obtain an overview of the missing values.

```python
# Calculate the total number of data for each column
total_values = len(olympics_data)

# Calculate the number of missing values for each column
missing_values_count = olympics_data.isnull().sum()

# Calculate the percentage of missing values for each column
missing_percentage = (missing_values_count / total_values) * 100

# Create a DataFrame to store the results
missing_data_summary = pd.DataFrame({
    'Column': missing_values_count.index,
    'Missing_Values': missing_values_count.values,
    'Percentage': np.round(missing_percentage.values, 2)
})

# Display the results
display(missing_data_summary)

# Number of columns containing missing values
num_columns_with_missing = len(missing_data_summary[missing_data_summary['Missing_Values'] > 0])

# Names of columns containing missing values
columns_with_missing_values = missing_data_summary[missing_data_summary['Missing_Values'] > 0]['Column'].tolist()

# Print the information
print("There are {} features containing missing values.".format(num_columns_with_missing))
print("These features are:")

# Iterate over columns with missing values
for column in columns_with_missing_values:
    # Find the percentage of missing values for the current column
    percentage_missing = missing_data_summary.loc[missing_data_summary['Column'] == column, 'Percentage'].iloc[0]
    # Print the column name and its corresponding percentage of missing values
    print("     Features '{}': {}% missing values".format(column, percentage_missing))
```

```python
# Group the data by 'ID' and 'Year' and keep the first row of each group (i.e., the first appearance of each athlete in each year)
grouped_data = olympics_data.groupby(['ID','Year']).first().reset_index()[['Year'] + columns_with_missing_values]
print('The number of rows in the grouped data is:', grouped_data.shape[0])

# Iterate through columns with missing values
for column in columns_with_missing_values:
    # Convert missing values to boolean (True if missing, False if not)
    grouped_data[column] = grouped_data[column].notnull()

# Calculate the percentage of non-missing values for each column grouped by year
grouped_data = grouped_data.groupby(by='Year').mean() * 100

# Display the first few rows of the grouped data
display(grouped_data.head())
```

```python
# Reset the index of grouped_data DataFrame
grouped_data.reset_index(inplace=True)

# Reshape the DataFrame using pd.melt to make it suitable for plotting
grouped_data = pd.melt(grouped_data, id_vars='Year', value_vars=columns_with_missing_values, value_name='Percentage', var_name='Features')
display(grouped_data.head())

# Create a bar plot with Seaborn
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x='Year', y='Percentage', hue='Features', data=grouped_data)
plt.xticks(rotation=90)  # Rotate year labels if necessary
plt.title('Percentage of Complete Data by Year')
plt.show()
```

Regarding age, only 4% of the total athlete data was missing. The distribution of these missing data seems to be generally balanced and not significant over the years, although it's unsurprising that we have more missing data on age in the past than now. As for height and weight characteristics, there appears to be an abrupt change in 1960 as we have this information for over 80% of our athletes. Conducting a thorough study would be interesting, particularly to determine if certain sports are more affected than others (for example, we know that weight in boxing is a selection factor, so we suspect there won't be missing data in this category).

1. Athletic Evolution 

```python
olympics_descriptives = olympics_data.groupby(['Year', 'Season', 'ID']).first().reset_index()[['Year', 'Season','Name', 'Sex', 'Age', 'Height', 'Weight', 'Sport']]
olympics_descriptives = olympics_descriptives[olympics_descriptives['Year'] >= 1960]
display(olympics_descriptives)
```

```python
display(olympics_descriptives.groupby(['Season', 'Year', 'Sex']).agg({'Age': ['min', 'max', 'mean', 'std'], 'Height': ['min', 'max', 'mean', 'std'], 'Weight': ['min', 'max', 'mean', 'std']}))
```

```python
# 1) Calculate the number of teams represented per year and season
teams_per_year_season = olympics_data.groupby(['Year', 'Season'])['Team'].nunique().reset_index()
teams_per_year_season.rename(columns={'Team': 'Number of Teams'}, inplace=True)

# Plotting the number of teams per year and season
plt.figure(figsize=(14, 7))
sns.barplot(x='Year', y='Number of Teams', hue='Season', data=teams_per_year_season)
plt.title('Number of Teams Represented Per Year and Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
medal_count = olympics_data.groupby(['Year', 'Season', 'Team', 'Medal'])[['Medal']].count()
medal_count.rename(columns={'Medal': 'Count'}, inplace=True)
medal_count.reset_index(inplace=True)
display(medal_count.head( ))
```

```python
participation = medal_count.groupby(['Season', 'Team'])[['Year']].nunique().sort_values(by = 'Year', ascending = False)
participation.rename(columns={'Year': 'Years Participated'}, inplace=True)
participation.reset_index(inplace=True)
display(participation)
```

```python
# Fusionner les deux DataFrames sur les colonnes 'Season' et 'Team'
medal_win_with_years = pd.merge(medal_count, participation, on=['Season', 'Team'], how='left')

# Afficher les rsultats
medal_win_with_years
```

```python
medal_win_pivot = medal_win_with_years.pivot_table(index=['Year', 'Season', 'Team', 'Years Participated'],
                                                   columns='Medal',
                                                   values='Count',
                                                   aggfunc='sum',
                                                   fill_value=0).reset_index()
medal_win_pivot
```

```python
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable 

# Separate the data for Summer and Winter Olympics
medal_win_summer = medal_win_with_years[medal_win_with_years['Season'] == 'Summer']
medal_win_winter = medal_win_with_years[medal_win_with_years['Season'] == 'Winter']

# Define the color mapping based on years participated
max_years_summer = medal_win_summer['Years Participated'].max()
max_years_winter = medal_win_winter['Years Participated'].max()
cmap = plt.get_cmap('coolwarm')
norm_summer = mcolors.Normalize(vmin=1, vmax=max_years_summer)
norm_winter = mcolors.Normalize(vmin=1, vmax=max_years_winter)

# Define medal types
medal_types = ['Gold', 'Silver', 'Bronze']

# Create a figure with subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 10), constrained_layout=True)

# Iterate over each medal type to create bar plots for both Summer and Winter Olympics
for i, medal in enumerate(medal_types):
    for j, (season_df, norm) in enumerate(zip([medal_win_summer, medal_win_winter], [norm_summer, norm_winter])):
        season = 'Summer' if j == 0 else 'Winter'
        # Filter for the specific type of medal and get top 10 teams
        medal_df = season_df[season_df['Medal'] == medal]
        top_teams = (medal_df.groupby('Team')
                              .agg({'Count': 'sum', 'Years Participated': 'first'})
                              .sort_values(by='Count', ascending=False)
                              .head(10))

        # Create color list based on the years participated
        color_values = [cmap(norm(years)) for years in top_teams['Years Participated']]
        
        # Create the bar plot
        sns.barplot(x='Count', y='Team', data=top_teams, ax=axes[i, j], palette=color_values)
        
        axes[i, j].set_title(f'Top 10 {medal} Medal Winning Teams in {season} Olympics')
        axes[i, j].set_xlabel('Medal Count')
        axes[i, j].set_ylabel('')

        # Add annotations for the number of participations
        for index, row in top_teams.iterrows():
            axes[i, j].text(row['Count'], index, f"{row['Years Participated']}", 
                             va='center', ha='left', color='black')

# Create a colorbar
sm = ScalarMappable(cmap=cmap, norm=norm_summer)
sm.set_array([])
cbar = fig.colorbar(sm, ax=axes[:, 1], location='right', aspect=10)
cbar.set_label('Number of participations')

# Show the plot
plt.show()
```
